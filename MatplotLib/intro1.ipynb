```python
import numpy as np
import matplotlib.pyplot as plt
```

Basic plotting

```python
x = np.linspace(0, 15, 100)
y = np.sin(x) + 0.01*np.random.randint(len(x))
plt.plot(x, y)
```

Beautification

```python
plt.figure(figsize=(10,3))
plt.plot(x, y, 'o--', color='black', lw=0.7, ms=5)
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
```

```python
plt.figure(figsize=(10,3))
plt.plot(x, y, 'o--', color='black', lw=0.7, ms=5, label="Component1")
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.legend(loc="upper right", fontsize=10)
```

#plotting two curves together

```python
x2 = np.linspace(0, 15, 100)
y2 = np.sin(x2)
plt.figure(figsize=(10, 3))
plt.plot(x, y, 'o', color='black', lw=0.7, ms=5, label="Component1")
plt.plot(x2, y2, color="purple", lw=0.7, ms=5, label="Component2")
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.legend(loc="upper right", fontsize=10)
```

#x and y limits

```python
x2 = np.linspace(0, 15, 100)
y2 = np.sin(x2)
plt.figure(figsize=(10, 3))
plt.plot(x, y, 'o', color='black', lw=0.7, ms=5, label="Data")
plt.plot(x2, y2, color="purple", lw=0.7, ms=5, label="Theory")
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.legend(loc="upper right", fontsize=10, ncol=2)
```

##Histogram##

```python
res = np.random.randn(1000)*0.2 + 0.4
```

Quick Histogram

```python
plt.hist(res)
```

```python
plt.figure(figsize=(8, 3))
plt.hist(res, density=True)
plt.show()
```

The below is difficult to see

```python
res2 = np.random.randn(1000)*0.2 + 0.4
plt.figure(figsize=(8, 3))
plt.hist(res, density=True)
plt.hist(res2, density=True)
plt.show()
```

Hence, we use, histtype='step'

```python
plt.figure(figsize=(8, 3))
plt.hist(res, density=True, histtype='step')
plt.hist(res2, density=True, histtype='step')
plt.show()
```
