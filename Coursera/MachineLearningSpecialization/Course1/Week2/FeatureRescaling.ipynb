# Feature Scaling

*Features and parameters*

This is when the feature is large a good model will choose a relatively smal parameter (weight)

This is related to gradient descent, as the countour plots in this case will form ellipses and prove to be shorter one side and longer on the other side, *since the countours are so tall and skinny the alues may bounce around a lot before reaching the minimum value*

hence we scale the feature such that the features scale from 0 to 1, and thus the countour plots will look more like circles, and hence easire to choose a path.

# How to scale?

One way is to scale using dividing by the maximum...

```python
# sq feet, bedrooms
x_train = np.array([[2000, 5], [300, 2], [400, 3],
                   [700, 2], [800, 4], [600, 2]])

# prices
y_train = np.array([500, 100, 250, 300, 400, 300])
```

```python
X_train = np.transpose(x_train)
plt.scatter(X_train[0], X_train[1], y_train, c='b')
```

Normalising

```python
plt.scatter(X_train[0]/np.max(X_train[0]), X_train[1]/np.max(X_train[1]), y_train, c='r')
```

# Mean Normalisation

Let the mean of the training feature $x_1$ be $\mu_1$
Then, 
$$x_1 = \frac{x_1-\mu_1}{max-min}$$

```python
Mean_1 = np.mean(X_train[0])
X__train = np.zeros((2, 6))
X__train[0] = (X_train[0]-Mean_1)/(np.max(X_train[0])-np.min(X_train[0]))
X__train[0]
```

```python
Mean_2 = np.mean(X_train[1])
X__train[1] = (X_train[1]-Mean_2)/(np.max(X_train[1])-np.min(X_train[1]))
X__train[1]
```

```python
plt.scatter(X__train[0], X__train[1], y_train, c='b')
```
