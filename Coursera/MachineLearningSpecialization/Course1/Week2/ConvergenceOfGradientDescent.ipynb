# Gradient Descent Convergence

The Objective is obviously to find the minimum of the cost function, so it is a good practice to plot the function J and see if the minimum is actually decreasing, For example...

```python
x = np.linspace(1, 40, 100)
y = lambda x: 1/x
plt.ylabel("Cost Function J(w, b)")
plt.xlabel("#Iterations")
plt.title("Learning Curve")
plt.plot(x, y(x))
```

If the value of $\alpha$ is chosed wrongly there is a chance for this function to become increasing and thus not converge

# Automatic Convergence test

Let $\epsilon$ be a small number like $10^{-3}$ <br>
If $J(\vec{w}, b)$ decreases by less than $\epsilon$ then we may stop the iterations <br>

But finding such epsilon can be a bit difficult and thus it is better to rely on the above done graphical test
