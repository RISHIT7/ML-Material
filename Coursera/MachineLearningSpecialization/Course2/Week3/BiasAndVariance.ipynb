## Bias and Variance

Remember Overfitting == High bias <br>
Underfitting == High Variance

#### High bias
1) $J_{train}$ is high
2) $J_{cv}$ is high

#### High Variance
1) $J_{train}$ is low
2) $J_{cv}$ is high

#### High Variance and High bias
1) $J_{train}$ will be very high
2) $J_{cv}$ >> $J_{train}$

## Learning Curves

In a general sense as the training set sizes grow the learning curves geenrally look like...

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y_1 = 1-np.exp(-x/5)
y_2 = 2.09-y_1
plt.xlabel('$m_{train}$')
plt.ylabel('error')
plt.plot(x, y_1, c='blue', label = 'training set error')
plt.plot(x, y_2, c='purple', label = 'cross validation error')
plt.legend()
```

### Incase of High bias the graphs will look like

```python
x = np.linspace(0, 10, 100)
y_1 = 1-np.exp(-5*x)
y_2 = 2.09-y_1
plt.xlabel('$m_{train}$')
plt.ylabel('error')
plt.plot(x, y_1, c='blue', label = 'training set error')
plt.plot(x, y_2, c='purple', label = 'cross validation error')
plt.legend()
```

If a learning algorithms has more bias then a model, no matter how many training data you get, will not help, and will never dip down to a human level performance

### Incase of High Variance the graphs will look like

```python
x = np.linspace(0, 10, 100)
y_1 = 1-np.exp(-x)
y_2 = 4-y_1
plt.xlabel('$m_{train}$')
plt.ylabel('error')
plt.plot(x, y_1, c='blue', label = 'training set error')
plt.plot(x, y_2, c='purple', label = 'cross validation error')
plt.legend()
```

Now in this case if we get a lot of training data then it will help as the training error will go up slightly but th cross validation error will also come down, and be eith close to or just little higher then the human level base line performance
