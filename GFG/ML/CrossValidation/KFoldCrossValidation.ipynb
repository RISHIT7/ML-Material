## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax

from sklearn import datasets
```

## Loading Data

```python
data = datasets.load_breast_cancer()
data.feature_names
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
df.head()
```

```python
df.iloc[:, (df.isnull().sum() != 0).values]
```

```python
x = np.array(df.iloc[:, :-1])
y = np.array(df.iloc[:, -1])
features = df.columns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
print (x_train.shape, y_train.shape)
print (x_test.shape, y_test.shape)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(100)
model = model.fit(x_train, y_train)

predictions = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, predictions)
```

## Storing the model

```python
import pickle as pkl
with open('./model.pickle', 'wb') as file:
    pkl.dump(model, file)
file.close()
```

## K-Fold Cross Validation

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits = 5, shuffle = False)

for train_index, test_index in kfold.split(x):
    print("Train - \n", train_index, "\nTest - \n", test_index)
    print("=======\n")
```

```python
rf1= RandomForestRegressor()

n_est = [100, 200, 500]

for n_estimator in n_est:       
    error = []
    for train_index, test_index in kfold.split(x):
        
        x_train, x_test = x[train_index], x[test_index]
        y_train, y_test = y[train_index], y[test_index]
    
        rf1 = RandomForestRegressor(n_estimator)

        rf1 = rf1.fit(x_train, y_train)

        prediction = rf1.predict(x_test)
        
        error.append(mean_squared_error(y_test, prediction))
    print("Results for n_estimator: ", n_estimator)
    print(error)
    print("Average Error = ", sum(error)/len(error))
```
