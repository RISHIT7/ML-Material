```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.isnull().sum()
```

## Distplot

```python
sns.displot(df['Age'], kde = True, color = 'blue')
```

```python
sns.displot(df['Cholesterol'], kde = True)
```

## Violin Plot

```python
sns.violinplot(x = df['Age'])
```

```python
sns.violinplot(y = df['Sex'], x = df['HeartDisease'])
```

```python
sns.violinplot(y = df['Age'], x = df['HeartDisease'])
```

```python
sns.violinplot(y = df['Cholesterol'], x = df['HeartDisease'])
```

## Correlation (Heat Map)

```python
num_df = df.drop(['ST_Slope', 'Sex', 'ChestPainType', 'ExerciseAngina', 'RestingECG'], axis = 1)
num_df.head()
```

```python
num_df.corr()
```

```python
sns.heatmap(num_df.corr())
```

## Correlation(JointPlot)

```python
sns.jointplot(x = 'Age' , y = 'MaxHR', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'Cholesterol' , y = 'MaxHR', data = df, kind = 'reg')
```

```python
sns.pairplot(data = df)
```
